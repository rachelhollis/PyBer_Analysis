```python
%matplotlib inline

# import matplotlib 
# pyplot for LINE CHARTS
import matplotlib.pyplot as plt
```

```python
savings = [0, 100, 150, 200, 300, 500, 300, 350, 500, 600, 750, 1000]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May','Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
```

```python
# create chart 
plt.plot(months, savings)

# Add labels
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Saving Account Results')

# shows the current plot
#ALWAYS THE FINAL PART OF CODE
# plt.show()
```

```python
# create a line chart using Object-Oriented interface

# create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(months, savings)
```

```python
#FOLLOWING MODULE
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#MATLAB APPROACH
# Create the plot and add a label for the legend.
# peep marker, color, linewidth, and label
plt.plot(x_axis, y_axis, marker ='*', color='blue', linewidth=2, label='Boston')

# create labels for x and y axis
plt.xlabel('Date')
plt.ylabel('Fare ($)')

# determine the y ranges
plt.ylim(0,45)
# determine x range practice
#plt.xlim(0,5)

# Create a title
plt.title('PyBer Fare By Month')

# add the legend
plt.legend()
```

```python
# OBJECT ORIENTED APPROACH

# create plot
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker='*', color='green', linewidth='2', label = 'Boston')

# label x-axis, y-axis and title
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_title('PyBer Fare By Month')

# add the legend
ax.legend()

#add grid
ax.grid()
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, 
           x_axis, 
           c='skyblue', 
           edgecolor='black', 
           linewidth=2,
           alpha=.2, 
           s = [i*5 for i in y_axis], 
           label='Boston')

# set labels
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_title('PyBer Fare By Month')

ax.set_xlim(0,50)
ax.invert_yaxis()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.pie(y_axis, 
        labels=x_axis)
plt.show()
```

```python
# explode the values
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, 
        explode=explode_values, 
        labels=x_axis, 
        autopct='%.1f%%')
plt.show()
```

```python
# increase the size of the chart, assign colors
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", 
          "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.subplots(figsize=(8, 8)) # this inc the size of the chart

plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(y_axis,
       labels=x_axis)
plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", 
          "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, .2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,
      labels=x_axis,
       explode=explode_values,
       shadow = 'True',
       autopct = '%.1f%%',
       startangle = 270,
       counterclock = 'True'
      )
ax.invert_yaxis()
plt.show()
```
